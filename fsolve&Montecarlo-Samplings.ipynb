When we have $n$ non-linear equations with $x_i,..., x_n$ unknown variables, we have a multivariable system of nonlinear equations with dimensional compatibility which can be expressed as: <br>
$f_1(x_1,x_2,...,x_n) = 0$ <br>
$f_2(x_1,x_2,...,x_n) = 0$ <br>
$\vdots$ <br>
$f_n(x_1,x_2,...,x_n) = 0$ <br>

One of the most common ways to solve it is with Newthon's method originated from the root of the first two terms of the Taylor series expansion:<br>
$f(x) = f(x_n) + f'(x_n)(x-x_n) + \frac{f''(x_n)}{2!}(x-x_n)^2 + \cdots$ <br>
$f(x) = f(x_n) + f'(x_n)(x-x_n)$ = 0 <br>
$x = x_n - \frac{f(x_n)}{f'(x_n)}$ <br><br>

Generalizing for $n$ variables, we have: <br>
$x_{n+1} = x_n - \frac{f(x_i)}{f'(x_i)}$ <br><br>

From Linear Algebra we know the Jacobian matrix is obtained by taking the first derivative of each function with respect to each variable. <br><br>

$x_{n+1} = x_n - \frac{1}{f'(x_i)} f(x_i)$<br>

$J(x) = \begin{bmatrix}
\frac{\partial f_1}{\partial x_1} & \frac{\partial f_1}{\partial x_2} & \cdots & \frac{\partial f_1}{\partial x_n} \\
\frac{\partial f_2}{\partial x_1} & \frac{\partial f_2}{\partial x_2} & \cdots & \frac{\partial f_2}{\partial x_n} \\
\vdots & \vdots & \ddots & \vdots \\
\frac{\partial f_n}{\partial x_1} & \frac{\partial f_n}{\partial x_2} & \cdots & \frac{\partial f_n}{\partial x_n} \\
\end{bmatrix}$ 

and rewrite the equation with $x_n$ and $f(x_i)$ as vectors:<br>
$x_n = \begin{bmatrix} x_1 \\ x_2 \\ \vdots \\ x_n \end{bmatrix}$ and 
$f(x_i) = \begin{bmatrix} f_1(x_1,x_2,...,x_n) \\ f_2(x_1,x_2,...,x_n) \\ \vdots \\ f_n(x_1,x_2,...,x_n) \end{bmatrix}$ <br><br><br>

So we have an iteration of the Newton-Raphson method for a multivariate system of nonlinear equations:<br>

$x_{n+1}$ = $x_n = \begin{bmatrix} x_1 \\ x_2 \\ \vdots \\ x_n \end{bmatrix}$ - 
$\begin{bmatrix} \frac{\partial f_1}{\partial x_1} & \frac{\partial f_1}{\partial x_2} & \cdots & \frac{\partial f_1}{\partial x_n} \\ \frac{\partial f_2}{\partial x_1} & \frac{\partial f_2}{\partial x_2} & \cdots & \frac{\partial f_2}{\partial x_n} \\ \vdots & \vdots & \ddots & \vdots \\
\frac{\partial f_n}{\partial x_1} & \frac{\partial f_n}{\partial x_2} & \cdots & \frac{\partial f_n}{\partial x_n} \\
\end{bmatrix}^{-1}$ $\cdot$ $\begin{bmatrix} f_1(x_1,x_2,...,x_n) \\ f_2(x_1,x_2,...,x_n) \\ \vdots \\ f_n(x_1,x_2,...,x_n) \end{bmatrix}$  

To start iterations an initial guess must be made and with it obtain $x_1$, then $x_2$ and so on..<br>

Convergence occurs when the change in $x_{n+1}$ from $x_{n}$ is $\epsilon$ (close to 0). 

There are other numerical methods to solve multivariate non linear systems but Newton is not so hard, very common and tends to be fast. 

In Python, we can use the `fsolve` function from `scipy.optimize` library to estimate their values.<br>

`fsolve` is a function that finds the roots of a system of non-linear equations and takes as arguments:
- The function $f(x)$ where $x = y, w, z$, etc.
- The initial guesses on $x_0$ for $x_n$.

`e.g: fsolve:` $\bigg[f(x_i), (x_{1_{g}},x_{2_{g}},...,x_{n_{g}})\bigg]$ <br>

So we have an iteration of the Newton-Raphson method for a multivariate system of nonlinear equations:<br>

##### *Beta distribution*

1. Obtaining parameters of distribution $x \sim \text{B}(\alpha, \beta).$ <br>
- Plot of $f(x)$ and $F(x)$ with $\alpha$ and $\beta$. <br>
- Probabilities for $f(x)$ and $F(x)$ with $\alpha$ and $\beta$. <br>

$f(x)$ of a $\beta$ distribution is given by: 

$f(x, a, b)=\frac{\Gamma(a+b) x^{a-1}(1-x)^{b-1}}{\Gamma(a) \Gamma(b)}$

and takes $\forall$ $a,b\in[0,1]$ as values so our initial guesses must be in that range and use the Gamma prob. densityfunction.

We can construct our system of linear equations as:

$\begin{bmatrix}
\frac{40\beta+49\alpha}{\alpha + \beta}-44\\
\frac{81 \alpha \beta}{(\alpha+\beta)^2(1+\alpha+\beta)}-9\\
\end{bmatrix}$
$\begin{bmatrix}
\alpha\\
\beta\\
\end{bmatrix}$

```python
#Libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats as st
from scipy.optimize import fsolve
from scipy.integrate import quad
%matplotlib inline
```

```python
#f(xi)
def fxi(Beta):
    alpha,beta=Beta
    return (((40*beta + 49*alpha)/(alpha + beta)) -44, ((alpha * beta*(9)**2)/((alpha + beta)**2*(1+alpha + beta)))-9)

#fsolve(fxi,(a_guess, b_guess))
try:
    alpha,beta=fsolve(fxi,(1,1))
    print("Alpha:",alpha, "Beta:",beta)

except:
    print("No solution found")
```

Now that we have obtained our parameters we can use them to plot the probability density function and the cumulative of the Beta distribution.

```python
#limits for xi
x=np.arange(40,49,0.01)

#Density functions
y=st.beta.pdf(x,alpha,beta, loc=40, scale=9)
Y=st.beta.cdf(x,alpha,beta, loc=40, scale=9)

#Plot of the density functions
plt.style.use('dark_background')
plt.figure(figsize = (20, 12))
plt.plot(x,y, linestyle='dotted', color='white')
plt.plot(x,Y, color='green')

#Style of the plot
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.title('β $f(x, a, b)$ and $F(x, a, b)$ density functions in $(a,b) \in [0,1]$',fontsize=20)
plt.xlabel('x',fontsize=14)
plt.ylabel('Probability',fontsize=14)
plt.legend(['Probability: f(x, a, b)','Accumulated: F(x, a, b)'],fontsize=14)
#Probabilities
df_beta = pd.DataFrame({'x':x, 'f(x)':st.beta.pdf(x,alpha,beta,loc=40,scale=9), 'F(x)':st.beta.cdf(x,alpha,beta,loc=40,scale=9),})

plt.show()

df_beta
```

##### *Triangular distribution*

- Plot $f(x)$ and $F(x)$. <br>
- Probabilities for $f(x)$ and $F(x)$. <br>

The Triangular probability density function is given by:
$f(x) = \begin{cases}
\frac{2(x-a)}{(b-a)(c-a)} & \text{for } a \leq x \leq c, \\
\frac{2(b-x)}{(b-a)(b-c)} & \text{for } c \leq x \leq b, \\
0 & \text{otherwise.}
\end{cases}$
and takes as arguments:
- $a$ = lower limit.
- $b$ = upper limit.
- $c$ = mode.<br>

an example of how to use is:<br>

`st.triang.pdf` $\bigg(x, c, loc=a, scale=b-a\bigg)$ 

```python
c = .44
x = np.linspace(40,49,100)
t = st.triang.pdf(x,c,loc=40,scale=9)
T = st.triang.cdf(x,c,loc=40,scale=9)
# graph
plt.style.use('dark_background')
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (20, 8))

ax1.plot(x,t, color='white')
ax2.plot(x,T, color='green')

#Make a shared title for both functions
fig.suptitle('Triangular f(x, c, a, b) and $F(x, c, a, b)$ density functions in $(a,b) \in [0,1]$',fontsize=20)
#Style of the plot
ax1.set_xlabel('x',fontsize=14)
ax1.set_ylabel('Probability',fontsize=14)
ax1.legend(['Probability: f(x, c, a, b)','Accumulated: F(x, c)'],fontsize=14)
ax1.grid(color='gray', linestyle='--', linewidth=0.5)

ax2.set_xlabel('x',fontsize=14)
ax2.set_ylabel('Probability',fontsize=14)
ax2.legend(['Probability: f(x, c, a, b)','Accumulated: F(x, c)'],fontsize=14)
ax2.grid(color='gray', linestyle='--', linewidth=0.5)

plt.show()

df2_triang = pd.DataFrame({'x':x, 'f(x)':t, 'F(x)':T})
df2_triang
```

<font color= 'orange'> Raw & Stratified Montecarlo Sampling: <font>

Montecarlo methods are an attractive choice as Machine Learning methods in order to sample normalized density functions. <br>

$V$ will be the volume of the integral on a prob. density function $f(x)$ with a normalized domain.

$$V = \int_{a}^{b} f(x) dx$$

First, $N$ random variables $x_{i} \in [a,b] \sim U(a,b)$ must be generated.<br>

*Note: Uniform Distribution is used for illustration purposes in this example.*<br> 
Any $f(x)$ given or obtained with known variables or datasets can be used. <br><br>

The prob. density function $f(x)$ for a continous uniform distribution is:

\begin{equation}
f(x) = \begin{cases}
\frac{1}{b-a} & \text{if } a \leq x \leq b\\
0 & \text{otherwise}
\end{cases}
\end{equation}

We must consider the property of expected values and the mathematical definition of the mean:<br>

$$E[X] = \int_{a}^{b} x f(x) dx$$

Then, the mean of $x_{i}$ in $V$ is:

$$F = \frac{V}{N} \sum_{i=0}^{N-1} f\left(x_i\right)$$

if we make $N = N-1$ we can rewrite the equation with the solution of the integral as:
$$F = \frac{b-a}{N-1} \sum_{i=0}^N f\left(x_i\right)$$

Montecarlo estimation for $N$ random variables $x_{i} \in$ $[a,b] \sim U(a,b)$ include the error estimator $\xi$.<br>
Assuming this, we obtain the estimator with the solution of the integral $\forall N$ and express it as $\hat{F}^{N}$ $\forall \xi \in$ $(0,1]$ so $\rightarrow$ $a = b$ can be assumed.

We can finally rewrite the estimator $\hat{F}^{N}$ as:

$$\hat{F}^{N} = \frac{b-a}{N-1} \sum_{i=0}^{N} f\bigg[a + \xi(b-a)\bigg]$$

```python
a,b = 0,1 #Integration interval
f= lambda x: 18*x**3+24*x**2+13*x+3  #Function to integrate
V = quad(f, a, b)[0] #Exact value of the integral
N = np.logspace(1,7,7,dtype=int) # N terms, in logarithmic scale (1, 10, 100, 1000, …, 10,000,000)
U = list(map(lambda N:np.random.uniform(a,b,N),N)) #Random numbers for the montecarlo method.


#Montecarlo integration function per se, based on random values.
def int_montecarlo(f:'Function to integrate',
                   a:'Lower limit of the integral',
                   b:'Upper limit of the integral',
                   U:'Sample of numbers U~[a,b]'):
    return (b-a)/len(U)*np.sum(f(U))

def stratification(N: 'N strata to create',
                   a:'Lower limit of the integral',
                   b:'Upper limit of the integral'):
    '''Function that receives and creates N number of strata for the stratification method:
    Divide the population into strata and generate a number of observations in each.
    proportional to its probability in order to ensure tha domain is covered and convergence is accelerated.
   '''
    u= np.random.uniform(a,b,N) #generate B random values ​​with uniform distribution on interval [a,b]
    i=np.arange(0,N) # variable iterable over the n layers created
    v = (u+i)/N #stratification
    return v


def approx_MC_EST(f:'Function to integrate',
                  a:'Lower limit of the integral',
                  b:'Upper limit of the integral',
                  N:'Number of terms'):
    '''Function that receives a function, the limits of the integral, and the number of terms to be used in the
    Monte Carlo method and the stratification method. It returns the value of the integral and the relative error
    of the Monte Carlo method and the stratification method.
    '''

    U2 = list(map(lambda N:stratification(N,a,b),N)) #Random numbers for the stratification method.
    M = [int_montecarlo(f, a, b, U) for U in U] #Application of the MonteCarlo method for the u1 generated numbers.
    EST = [int_montecarlo(f, a, b, U) for U in U2] #Application of the MonteCarlo method for the u2 stratified variables.
    # Dataframe fill
    df_cv = pd.DataFrame(index=N,columns=['Crude Monte Carlo', 'Relative MC_Error %', 'Stratified Sampling',
                                        'Relative_error M_est %' ], dtype='float') # Creation of DataFrame columns
    df_cv.index.name = "Terms" #Name df index.
    df_cv.loc[N,"Crude Monte Carlo"] = M #Raw Monte Carlo method on DataFrame over N terms.
    df_cv.loc[N,"Relative MC_Error %"] = np.abs(df_cv.loc[N,"Crude Monte Carlo"]-V)*100/V #Relative error of the Monte Carlo method
    df_cv.loc[N,"Stratified Sampling"] = EST #Stratification Method
    df_cv.loc[N,"Relative_error M_est %"] = np.abs(df_cv.loc[N,"Stratified Sampling"]-V)*100/V #Relative error of the stratification method
    #Plot
    plt.figure(figsize = (20, 12))
    plt.xscale('log') #Logarithmic scale on the x axis   
    plt.plot(df_cv.index,df_cv['Crude Monte Carlo'],marker='o',linestyle='--')
    plt.plot(df_cv.index,df_cv['Stratified Sampling'],marker='o',linestyle='--')
    # Plot Style
    plt.style.use('dark_background')
    plt.grid(linestyle='--')
    plt.title('Monte Carlo with Raw-Stratified Samplings:',fontsize=18)
    plt.xlabel('x',fontsize=14)
    plt.ylabel('Probability',fontsize=14)
    plt.legend(['Monte Carlo Raw Samplings','Monte Carlo Stratified Samplings'],fontsize=14)
    return df_cv, plt.show()

approx_MC_EST(f, a, b, N)[0]
```

By dividing Montecarlo samplings into strata we can see how convergence speeds are faster and more accurate.
